## Best Model for ReLU activation

```python
best_net,history = training(250,0.001,0.001,'relu',verbose = True)
```

```python
import scipy.io as scio
```

```python
def data_load():
    mnist = scio.loadmat('mnist-original.mat')
    X = mnist['data'].astype('float64')
    y = mnist['label'].astype('int64').reshape(-1)
    X_train = X[:,:60000].T
    X_test = X[:,60000:].T
    y_train = y[:60000]
    y_test = y[60000:]
    return X_train,y_train,X_test,y_test
def accuracy(y_pred:np.array,y_truth:np.array):
    return (y_pred == y_truth).sum() / y_pred.shape[0]
```

```python
X_train,y_train,X_test,y_test = data_load()
pred = best_net.predict(X_test)
test_acc = accuracy(pred,y_test)
```

```python
print(test_acc)
```

```python
import matplotlib.pyplot as plt

train_loss = history['train loss']
val_loss = history['val loss']
test_loss = history['test loss']
plt.figure(figsize = (10,5),dpi = 224)
plt.plot(np.arange(1,len(train_loss)+1),train_loss,label = 'train loss')
plt.plot(np.arange(1,len(train_loss)+1),val_loss,label = 'val loss')
plt.plot(np.arange(1,len(train_loss)+1),test_loss,label = 'test loss')
plt.title("Loss versus epoch")
plt.legend()
plt.show()
```

```python
train_acc = history['train acc']
val_acc = history['val acc']
test_acc = history['test acc']
plt.figure(figsize = (10,5),dpi = 224)
plt.plot(np.arange(1,len(train_acc)+1),train_acc,label = 'train accuracy')
plt.plot(np.arange(1,len(train_acc)+1),val_acc,label = 'val accuracy')
plt.plot(np.arange(1,len(train_acc)+1),test_acc,label = 'test accuracy')
plt.title("Accuracy versus epoch")
plt.legend()
plt.show()
```
